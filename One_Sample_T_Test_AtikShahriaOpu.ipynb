```python
# Statistical Analysis: One-Sample t-Test
# Analysis for Student ID: 79
# Course Assignment: Testing Population Mean Hypothesis

# Required libraries
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle
import pandas as pd

# Configure plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

def main():
    """
    Main function to execute the complete statistical analysis workflow
    """

    # Configuration and setup
    STUDENT_ID = 79
    SAMPLE_SIZE = 12
    NULL_HYPOTHESIS_MEAN = 65
    SIGNIFICANCE_LEVEL = 0.05
    CONFIDENCE_LEVEL = 0.95

    print("‚ïî" + "="*60 + "‚ïó")
    print("‚ïë" + " STATISTICAL HYPOTHESIS TESTING ANALYSIS ".center(60) + "‚ïë")
    print("‚ïë" + f" Student ID: {STUDENT_ID} ".center(60) + "‚ïë")
    print("‚ïö" + "="*60 + "‚ïù")

    # Phase 1: Data Generation and Exploration
    dataset = generate_exam_scores(STUDENT_ID, SAMPLE_SIZE)
    descriptive_stats = compute_descriptive_statistics(dataset)

    # Phase 2: Hypothesis Framework
    display_hypothesis_framework(NULL_HYPOTHESIS_MEAN, SIGNIFICANCE_LEVEL)

    # Phase 3: Statistical Testing
    test_results = conduct_t_test(dataset, NULL_HYPOTHESIS_MEAN)
    confidence_bounds = calculate_confidence_interval(
        dataset, CONFIDENCE_LEVEL
    )

    # Phase 4: Decision Making
    statistical_decision = make_statistical_decision(
        test_results, SIGNIFICANCE_LEVEL, NULL_HYPOTHESIS_MEAN, confidence_bounds
    )

    # Phase 5: Comprehensive Visualization
    create_analysis_dashboard(
        dataset, descriptive_stats, test_results,
        confidence_bounds, NULL_HYPOTHESIS_MEAN, SIGNIFICANCE_LEVEL
    )

    # Phase 6: Final Report
    generate_comprehensive_report(
        descriptive_stats, test_results, confidence_bounds,
        statistical_decision, NULL_HYPOTHESIS_MEAN
    )

def generate_exam_scores(student_id, sample_size):
    """
    Generate exam score dataset using the specified algorithm

    Algorithm: scores = 60 + random_int(0,14) + (student_id % 5)
    """

    # Set seed based on student ID for reproducible results
    np.random.seed(student_id)

    k_value = student_id
    modulo_adjustment = k_value % 5
    base_score = 60

    print(f"\nüìä DATASET GENERATION PROCESS")
    print(f"‚îå‚îÄ Student ID (k): {k_value}")
    print(f"‚îú‚îÄ Modulo calculation: {k_value} % 5 = {modulo_adjustment}")
    print(f"‚îú‚îÄ Base score: {base_score}")
    print(f"‚îî‚îÄ Formula: {base_score} + randint(0,15) + {modulo_adjustment}")

    # Generate random components
    random_additions = np.random.randint(0, 15, size=sample_size)

    # Calculate final scores
    exam_scores = base_score + random_additions + modulo_adjustment

    print(f"\nüìã Generated Exam Scores:")
    print(f"Random components: {list(random_additions)}")
    print(f"Final scores: {list(exam_scores)}")

    return exam_scores

def compute_descriptive_statistics(data):
    """
    Calculate comprehensive descriptive statistics for the dataset
    """
    mode_result = stats.mode(data)
    # Check if the mode result's mode array is empty before accessing the first element
    mode_value = mode_result.mode[0] if len(mode_result.mode) > 0 else 'No unique mode'


    stats_dict = {
        'count': len(data),
        'mean': np.mean(data),
        'median': np.median(data),
        'mode': mode_value,
        'std_dev': np.std(data, ddof=1),
        'variance': np.var(data, ddof=1),
        'std_error': np.std(data, ddof=1) / np.sqrt(len(data)),
        'min_value': np.min(data),
        'max_value': np.max(data),
        'range': np.max(data) - np.min(data),
        'q1': np.percentile(data, 25),
        'q3': np.percentile(data, 75),
        'iqr': np.percentile(data, 75) - np.percentile(data, 25)
    }

    print(f"\nüìà DESCRIPTIVE STATISTICS SUMMARY")
    print(f"‚îå‚îÄ Sample Size: {stats_dict['count']}")
    print(f"‚îú‚îÄ Mean (xÃÑ): {stats_dict['mean']:.4f}")
    print(f"‚îú‚îÄ Median: {stats_dict['median']:.4f}")
    print(f"‚îú‚îÄ Standard Deviation (s): {stats_dict['std_dev']:.4f}")
    print(f"‚îú‚îÄ Standard Error (SE): {stats_dict['std_error']:.4f}")
    print(f"‚îú‚îÄ Range: [{stats_dict['min_value']}, {stats_dict['max_value']}]")
    print(f"‚îî‚îÄ Interquartile Range: {stats_dict['iqr']:.4f}")

    return stats_dict

def display_hypothesis_framework(null_mean, alpha):
    """
    Present the hypothesis testing framework clearly
    """

    print(f"\nüéØ HYPOTHESIS TESTING FRAMEWORK")
    print(f"‚îå‚îÄ Null Hypothesis (H‚ÇÄ): Œº = {null_mean}")
    print(f"‚îÇ  'The true population mean equals {null_mean}'")
    print(f"‚îú‚îÄ Alternative Hypothesis (H‚ÇÅ): Œº ‚â† {null_mean}")
    print(f"‚îÇ  'The true population mean does not equal {null_mean}'")
    print(f"‚îú‚îÄ Test Type: Two-tailed test")
    print(f"‚îú‚îÄ Significance Level (Œ±): {alpha}")
    print(f"‚îî‚îÄ Critical Region: |t| > t_{alpha/2}, df")

def conduct_t_test(data, hypothesized_mean):
    """
    Execute the one-sample t-test and return comprehensive results
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    degrees_freedom = sample_size - 1

    # Perform the t-test
    t_stat, p_val = stats.ttest_1samp(data, hypothesized_mean)

    # Calculate critical value
    critical_t = stats.t.ppf(0.975, degrees_freedom)  # Two-tailed test

    results = {
        'sample_mean': sample_mean,
        'hypothesized_mean': hypothesized_mean,
        't_statistic': t_stat,
        'p_value': p_val,
        'degrees_freedom': degrees_freedom,
        'critical_value': critical_t,
        'sample_std': sample_std,
        'sample_size': sample_size
    }

    print(f"\nüßÆ T-TEST COMPUTATION RESULTS")
    print(f"‚îå‚îÄ Sample Mean (xÃÑ): {results['sample_mean']:.4f}")
    print(f"‚îú‚îÄ Hypothesized Mean (Œº‚ÇÄ): {results['hypothesized_mean']}")
    print(f"‚îú‚îÄ t-Statistic: {results['t_statistic']:.4f}")
    print(f"‚îú‚îÄ Degrees of Freedom: {results['degrees_freedom']}")
    print(f"‚îú‚îÄ p-value: {results['p_value']:.6f}")
    print(f"‚îî‚îÄ Critical t-value (¬±): {results['critical_value']:.4f}")

    return results

def calculate_confidence_interval(data, confidence_level):
    """
    Compute confidence interval for the sample mean
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    degrees_freedom = sample_size - 1
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate confidence interval
    ci_bounds = stats.t.interval(
        confidence_level,
        degrees_freedom,
        loc=sample_mean,
        scale=standard_error
    )

    margin_error = (ci_bounds[1] - ci_bounds[0]) / 2

    interval_info = {
        'confidence_level': confidence_level,
        'lower_bound': ci_bounds[0],
        'upper_bound': ci_bounds[1],
        'margin_of_error': margin_error,
        'width': ci_bounds[1] - ci_bounds[0]
    }

    print(f"\nüìä CONFIDENCE INTERVAL ANALYSIS")
    print(f"‚îå‚îÄ Confidence Level: {confidence_level*100}%")
    print(f"‚îú‚îÄ Lower Bound: {interval_info['lower_bound']:.4f}")
    print(f"‚îú‚îÄ Upper Bound: {interval_info['upper_bound']:.4f}")
    print(f"‚îú‚îÄ Margin of Error: ¬±{interval_info['margin_of_error']:.4f}")
    print(f"‚îî‚îÄ Interval Width: {interval_info['width']:.4f}")

    return interval_info

def make_statistical_decision(test_results, alpha, null_mean, ci_info):
    """
    Make and justify the statistical decision
    """

    p_value = test_results['p_value']
    t_stat = test_results['t_statistic']
    critical_val = test_results['critical_value']

    # Decision based on p-value
    reject_null_pvalue = p_value < alpha

    # Decision based on critical value
    reject_null_critical = abs(t_stat) > critical_val

    # Decision based on confidence interval
    null_in_ci = ci_info['lower_bound'] <= null_mean <= ci_info['upper_bound']

    decision_info = {
        'reject_null': reject_null_pvalue,
        'p_value_decision': reject_null_pvalue,
        'critical_value_decision': reject_null_critical,
        'ci_decision': not null_in_ci,
        'effect_size': (test_results['sample_mean'] - null_mean) / test_results['sample_std']
    }

    print(f"\n‚öñÔ∏è  STATISTICAL DECISION ANALYSIS")
    print(f"‚îå‚îÄ p-value Method: {'REJECT H‚ÇÄ' if reject_null_pvalue else 'FAIL TO REJECT H‚ÇÄ'}")
    print(f"‚îú‚îÄ Critical Value Method: {'REJECT H‚ÇÄ' if reject_null_critical else 'FAIL TO REJECT H‚ÇÄ'}")
    print(f"‚îú‚îÄ Confidence Interval Method: {'REJECT H‚ÇÄ' if not null_in_ci else 'FAIL TO REJECT H‚ÇÄ'}")
    print(f"‚îú‚îÄ Effect Size (Cohen's d): {decision_info['effect_size']:.4f}")

    if decision_info['reject_null']:
        print(f"‚îî‚îÄ CONCLUSION: We have sufficient evidence to reject H‚ÇÄ")
        print(f"   The sample provides statistically significant evidence that")
        print(f"   the population mean differs from {null_mean}.")
    else:
        print(f"‚îî‚îÄ CONCLUSION: We fail to reject H‚ÇÄ")
        print(f"   The sample does not provide sufficient evidence to conclude")
        print(f"   that the population mean differs from {null_mean}.")

    return decision_info

def create_analysis_dashboard(data, desc_stats, test_results, ci_info, null_mean, alpha):
    """
    Generate a comprehensive visualization dashboard
    """

    fig = plt.figure(figsize=(16, 12))

    # Plot 1: Data Distribution with Statistical Markers
    ax1 = plt.subplot(2, 3, 1)
    plt.hist(data, bins=6, alpha=0.7, color='skyblue', edgecolor='navy', linewidth=1.2)
    plt.axvline(desc_stats['mean'], color='red', linestyle='-', linewidth=3, label=f"Sample Mean: {desc_stats['mean']:.2f}")
    plt.axvline(null_mean, color='orange', linestyle='--', linewidth=3, label=f"H‚ÇÄ Mean: {null_mean}")
    plt.xlabel('Exam Scores')
    plt.ylabel('Frequency')
    plt.title('Distribution of Exam Scores', fontweight='bold', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot 2: Box Plot with Confidence Interval
    ax2 = plt.subplot(2, 3, 2)
    box = plt.boxplot(data, patch_artist=True, labels=['Exam Scores'])
    box['boxes'][0].set_facecolor('lightcoral')
    plt.axhline(null_mean, color='orange', linestyle='--', linewidth=2, label=f'H‚ÇÄ: Œº = {null_mean}')
    plt.axhline(ci_info['lower_bound'], color='green', linestyle=':', alpha=0.7, label='95% CI')
    plt.axhline(ci_info['upper_bound'], color='green', linestyle=':', alpha=0.7)
    plt.ylabel('Score Values')
    plt.title('Box Plot with Hypothesis Mean', fontweight='bold', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot 3: Mean Comparison with Error Bars
    ax3 = plt.subplot(2, 3, 3)
    categories = ['Sample\nMean', 'Hypothesized\nMean']
    means = [desc_stats['mean'], null_mean]
    errors = [desc_stats['std_error'], 0]
    colors = ['crimson', 'darkorange']

    bars = plt.bar(categories, means, yerr=errors, capsize=8, color=colors, alpha=0.8,
                   error_kw={'elinewidth': 3, 'capthick': 3})
    plt.ylabel('Mean Value')
    plt.title('Mean Comparison with Standard Error', fontweight='bold', fontsize=12)
    plt.grid(True, alpha=0.3, axis='y')

    # Add value labels on bars
    for i, (bar, mean) in enumerate(zip(bars, means)):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + errors[i] + 0.5,
                f'{mean:.2f}', ha='center', va='bottom', fontweight='bold')

    # Plot 4: t-Distribution with Test Statistic
    ax4 = plt.subplot(2, 3, 4)
    t_values = np.linspace(-4, 4, 1000)
    t_density = stats.t.pdf(t_values, test_results['degrees_freedom'])
    plt.plot(t_values, t_density, 'blue', linewidth=2, label=f't-distribution (df={test_results["degrees_freedom"]})')

    # Shade critical regions
    critical_val = test_results['critical_value']
    left_critical = t_values[t_values <= -critical_val]
    right_critical = t_values[t_values >= critical_val]

    plt.fill_between(left_critical, 0, stats.t.pdf(left_critical, test_results['degrees_freedom']),
                     color='red', alpha=0.3, label=f'Critical Regions (Œ±={alpha})')
    plt.fill_between(right_critical, 0, stats.t.pdf(right_critical, test_results['degrees_freedom']),
                     color='red', alpha=0.3)

    # Mark test statistic
    plt.axvline(test_results['t_statistic'], color='darkgreen', linestyle='-', linewidth=4,
                label=f't-stat: {test_results["t_statistic"]:.3f}')

    plt.xlabel('t-value')
    plt.ylabel('Probability Density')
    plt.title('t-Distribution and Test Statistic', fontweight='bold', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot 5: Confidence Interval Visualization
    ax5 = plt.subplot(2, 3, 5)
    plt.errorbar(1, desc_stats['mean'],
                yerr=[[desc_stats['mean'] - ci_info['lower_bound']],
                      [ci_info['upper_bound'] - desc_stats['mean']]],
                fmt='ro', markersize=12, capsize=15, capthick=4,
                linewidth=3, label='95% Confidence Interval')

    plt.axhline(null_mean, color='orange', linestyle='--', linewidth=3,
                label=f'Null Hypothesis (Œº = {null_mean})')

    plt.xlim(0.5, 1.5)
    plt.ylabel('Score Value')
    plt.title('Confidence Interval vs Null Hypothesis', fontweight='bold', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks([1], ['Sample'])

    # Plot 6: Statistical Summary
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')

    # Create summary text
    summary_text = f"""
STATISTICAL SUMMARY
{'='*30}
Sample Size: {test_results['sample_size']}
Sample Mean: {desc_stats['mean']:.4f}
Standard Deviation: {desc_stats['std_dev']:.4f}
Standard Error: {desc_stats['std_error']:.4f}

TEST RESULTS
{'='*30}
t-statistic: {test_results['t_statistic']:.4f}
p-value: {test_results['p_value']:.6f}
Critical t-value: ¬±{test_results['critical_value']:.4f}

DECISION
{'='*30}
Œ± = {alpha}
{'REJECT H‚ÇÄ' if test_results['p_value'] < alpha else 'FAIL TO REJECT H‚ÇÄ'}
    """

    plt.text(0.1, 0.9, summary_text, transform=ax6.transAxes, fontsize=11,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.suptitle('One-Sample t-Test Analysis Dashboard', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

def generate_comprehensive_report(desc_stats, test_results, ci_info, decision_info, null_mean):
    """
    Generate a detailed final report
    """

    print(f"\n" + "="*80)
    print("COMPREHENSIVE STATISTICAL ANALYSIS REPORT".center(80))
    print("="*80)

    # Executive Summary
    print(f"\nüìã EXECUTIVE SUMMARY")
    print(f"This analysis examined whether a sample of {test_results['sample_size']} exam scores")
    print(f"provides evidence that the population mean differs from {null_mean}.")

    decision_text = "REJECTED" if decision_info['reject_null'] else "NOT REJECTED"
    print(f"The null hypothesis was {decision_text} based on statistical evidence.")

    # Detailed Findings
    print(f"\nüîç DETAILED FINDINGS")

    # Data characteristics
    print(f"\nData Characteristics:")
    print(f"‚Ä¢ Sample mean: {desc_stats['mean']:.4f} (vs. hypothesized {null_mean})")
    print(f"‚Ä¢ Sample standard deviation: {desc_stats['std_dev']:.4f}")
    print(f"‚Ä¢ Standard error: {desc_stats['std_error']:.4f}")
    print(f"‚Ä¢ Score range: [{desc_stats['min_value']}, {desc_stats['max_value']}]")

    # Statistical test results
    print(f"\nHypothesis Test Results:")
    print(f"‚Ä¢ t-statistic: {test_results['t_statistic']:.4f}")
    print(f"‚Ä¢ Degrees of freedom: {test_results['degrees_freedom']}")
    print(f"‚Ä¢ p-value: {test_results['p_value']:.6f}")
    print(f"‚Ä¢ Critical value (two-tailed): ¬±{test_results['critical_value']:.4f}")

    # Effect size interpretation
    effect_size = abs(decision_info['effect_size'])
    if effect_size < 0.2:
        effect_interpretation = "negligible"
    elif effect_size < 0.5:
        effect_interpretation = "small"
    elif effect_size < 0.8:
        effect_interpretation = "medium"
    else:
        effect_interpretation = "large"

    print(f"‚Ä¢ Effect size (Cohen's d): {decision_info['effect_size']:.4f} ({effect_interpretation})")

    # Confidence interval
    print(f"\nConfidence Interval:")
    print(f"‚Ä¢ 95% CI: [{ci_info['lower_bound']:.4f}, {ci_info['upper_bound']:.4f}]")
    print(f"‚Ä¢ Margin of error: ¬±{ci_info['margin_of_error']:.4f}")

    contains_null = ci_info['lower_bound'] <= null_mean <= ci_info['upper_bound']
    print(f"‚Ä¢ Null hypothesis value ({null_mean}) is {'within' if contains_null else 'outside'} the CI")

    # Final interpretation
    print(f"\nüí° INTERPRETATION")
    if decision_info['reject_null']:
        print(f"The statistical analysis provides strong evidence (p = {test_results['p_value']:.6f})")
        print(f"to conclude that the population mean is significantly different from {null_mean}.")
        print(f"The sample mean of {desc_stats['mean']:.4f} represents a statistically")
        print(f"significant departure from the hypothesized value.")
    else:
        print(f"The statistical analysis does not provide sufficient evidence")
        print(f"(p = {test_results['p_value']:.6f}) to conclude that the population mean")
        print(f"differs from {null_mean}. The observed difference could reasonably")
        print(f"be due to random sampling variation.")

    print(f"\n" + "="*80)

# Execute the complete analysis
if __name__ == "__main__":
    main()
```
